## Soal 1
**Identifikasi Individu dengan Semua Proyek Selesai**

Buat fungsi bernama **find_individuals_with_all_completed_projects** yang
mengembalikan list nama individu yang semua proyeknya sudah selesai. Proyek
dianggap selesai jika semua tugas di dalamnya memiliki status "completed".

Contoh Output:\
['John Doe', 'Emily Davis']

```python
def find_individuals_with_all_completed_projects(data):
    hasil = []

    for person in data:
        if person['projects']:
            #print(person['projects'])
            for project in person['projects']:
                if project['tasks']:
                    #print(project['tasks'])
                    flag = True
                    for task in project['tasks']:
                        if task['status'] != 'completed':
                            flag = False
                    if flag == True:
                        hasil.append(person['name'])
    
    return hasil

find_individuals_with_all_completed_projects(data)
```

## Soal 2
**Gabungkan Tugas Berdasarkan Kota**

Buat fungsi bernama **combine_tasks_by_city** yang mengelompokkan semua tugas
berdasarkan kota dari alamat home individu. Setiap tugas harus menyertakan nama
individu, ID proyek, dan ID tugas. Hasilnya berupa dictionary dengan nama kota sebagai
kunci dan list tugas sebagai nilai.

Contoh Output:\
{\
"New York": [\
{"name": "John Doe", "project_id": "P001", "task_id": "T001"},\
{"name": "John Doe", "project_id": "P001", "task_id": "T002"}\
],\
"Los Angeles": [\
{"name": "Jane Smith", "project_id": "P002", "task_id": "T003"},\
{"name": "Jane Smith", "project_id": "P002", "task_id": "T004"}\
],\
...\
}

```python
def combine_tasks_by_city(data):
    hasil = {f"{person['contacts']['address']['home']['city']}":[] for person in data }

    for person in data:
        kota = person['contacts']['address']['home']['city']
        if person['projects']:
            #print(person['projects'])
            for project in person['projects']:
                if project['tasks']:
                    #print(project['tasks'])
                    for task in project['tasks']:
                        #print(f"{kota}\n")
                        hasil[kota].append({"name": person['name'], "project_id": project['project_id'], "task_id": task['task_id']})

    return hasil

combine_tasks_by_city(data)
```

## Soal 3

**Identifikasi Kandidat Pelatihan Manajemen**


Buatlah fungsi bernama *find_promotion_candidates* yang mengidentifikasi individu yang cocok untuk promosi berdasarkan kriteria berikut:
1. Pengalaman Kerja (Experience (Years)) harus kurang dari 3 tahun.
2. Skor Kinerja (Performance Score) harus di atas rata-rata skor kinerja dari semua individu.
3. Gaji (Salary) harus berada di kuartil bawah (25% terendah) dari semua data gaji.
4. Individu tersebut harus bekerja di departemen yang bukan "Admin" atau "Support".


Fungsi ini harus mengembalikan list berisi dictionary untuk setiap individu yang
memenuhi kriteria, dengan informasi berikut:\
● Nama\
● Departemen\
● Gaji\
● Skor Kinerja\
● Pengalaman Kerja

**Hints**:\
● Anda harus membaca file dalam bentuk pandas dataframe.\
● Gunakan metode statistik untuk menghitung rata-rata skor kinerja dan kuartil bawah untuk gaji.\
● Pastikan individu dari departemen "Admin" dan "Support" tidak termasuk.

**Contoh Output**:\
[\
{"Name": "Hannah", "Department": "Procurement", "Salary": 52000, "Performance Score": 89, "Experience (Years)": 1},\
{"Name": "Mia", "Department": "Training", "Salary": 53000, "Performance Score": 88, "Experience (Years)": 2},\
...\
]

```python
def find_promotion_candidates(data):
    mean = data.describe()['Performance Score']['mean'] #menyimpan rataan dari skor kinerja semua individu
    Q1 = data.describe()['Salary']['25%'] #menyimpan kuartil satu dari varibael Gaji dari semua individu

    df1 = data[data['Experience (Years)'] < 3] #Pengalaman Kerja (Experience (Years)) harus kurang dari 3 tahun.
    df2 = df1[df1['Performance Score'] >  mean] #Skor Kinerja (Performance Score) harus di atas rata-rata skor kinerja dari semua individu.
    df3 = df2[df2['Salary'] < Q1] #Gaji (Salary) harus berada di kuartil bawah (25% terendah) dari semua data gaji.
    df4 = df3[(df3['Department'] != 'Admin') & (df3['Department'] != 'Support')] #Individu tersebut harus bekerja di departemen yang bukan "Admin" atau "Support".
    df5 = df4[['Name', 'Department', 'Salary','Performance Score', 'Experience (Years)']] # memilih kunci untuk dictionary

    hasil1 = df5.transpose()
    hasil2 = hasil1.to_dict()
    hasil3 = hasil2.values()
    hasil4 = list(hasil3)
    return hasil4

find_promotion_candidates(df)
```

## Soal 4

**Identifikasi Individu Ideal untuk Promosi**


Buatlah fungsi bernama *find_promotion_candidates* yang mengidentifikasi individu yang cocok untuk promosi berdasarkan kriteria berikut:
1. Pengalaman Kerja (Experience (Years)) harus minimal 5 tahun.
2. Skor Kinerja (Performance Score) harus di atas 85.
3. Status Remote Worker (Remote Worker) harus False (tidak bekerja secara remote).
4. Gaji (Salary) saat ini harus kurang dari rata-rata gaji di departemennya.

Fungsi ini harus mengembalikan list berisi dictionary untuk setiap individu yang memenuhi kriteria, dengan informasi berikut:\
● Nama\
● Departemen\
● Gaji\
● Skor Kinerja\
● Pengalaman Kerja

**Hints**:\
● Anda harus membaca file dalam bentuk pandas dataframe.\
● Gunakan rata-rata gaji untuk setiap departemen sebagai pembanding.\
● Pastikan fungsi ini menangani logika yang kompleks dengan menggabungkan beberapa kolom dan kriteria.

**Contoh Output**:\
[\
{"Name": "Chris", "Department": "Support", "Salary": 60000, "Performance Score": 84, "Experience (Years)": 6},\
{"Name": "William", "Department": "Operations", "Salary": 59000, "Performance Score": 86, "Experience (Years)": 7},\
...\
]

```python
def find_promotion_candidates(data):
    hasil = []

    def rataan_gaji_departemen(nama_departemen):
        df_departmen_gaji = data[['Department', 'Salary']]
        rata_gaji_departemen = df_departmen_gaji.groupby('Department').mean()
        rataan = rata_gaji_departemen.loc[nama_departemen]['Salary']
        return rataan

    df1 = df[df['Experience (Years)'] >= 5] #Pengalaman Kerja (Experience (Years)) harus minimal 5 tahun.
    df2 = df1[df1['Performance Score'] > 85] #Skor Kinerja (Performance Score) harus di atas 85.
    df3 = df2[df2['Remote Worker'] == False] #Status Remote Worker (Remote Worker) harus False (tidak bekerja secara remote).
    df3 #Gaji (Salary) saat ini harus kurang dari rata-rata gaji di departemennya.

    df4 = list(df3.transpose().to_dict().values())
    for i in df4:
        #print(i['Salary'], rataan_gaji_departemen(i['Department']))
        if i['Salary'] < rataan_gaji_departemen(i['Department']):
            hasil.append({"Name": i['Name'], "Department": i['Department'], "Salary": i["Salary"], "Performance Score": i['Performance Score'], "Experience (Years)": i['Performance Score']})

    return hasil

find_promotion_candidates(df)
```

# Soal 5

**Sorting dengan Kunci Kustom**

Diberikan list tuple yang berisi nama dan nilai:\
students = [("Ali", 85), ("Budi", 92), ("Citra", 78), ("Dewi", 95), ("Eka", 88)]

Buatlah sebuah fungsi *sort_students* yang mengembalikan list yang diurutkan berdasarkan nilai secara **descending/ascending**.

**Contoh Penggunaan**:\
sort_students(students, ‘descending’)

**Output yang diharapkan**:\
[("Dewi", 95), ("Budi", 92), ("Eka", 88), ("Ali", 85), ("Citra", 78)]

```python
students = [("Ali", 85), ("Budi", 92), ("Citra", 78), ("Dewi", 95), ("Eka", 88)]

def sort_students1(students, par):
    if par == 'ascending':
        return sorted(students, key=lambda x:x[1], reverse=False)
    if par == 'descending':
        return sorted(students, key=lambda x:x[1], reverse=True)

def sort_students2(students, par):
    for i in range(len(students)):
        for j in range(i):
            if par == 'ascending':
                if students[j][1] > students[j+1][1]:
                    students[j], students[j+1] =  students[j+1], students[j]
            if par == 'descending':
                if students[j][1] < students[j+1][1]:
                    students[j], students[j+1] =  students[j+1], students[j]
    return students

sort_students2(students, 'ascending')
```
